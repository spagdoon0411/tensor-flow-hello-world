```python
import pandas as pd
import sklearn 
from sklearn import compose
import numpy as np
import tensorflow as tf
from keras import layers
```

```python
dataset = pd.read_csv("./who_life_expect_dataset.csv")
display(dataset.head())
```

```python
dataset = dataset.drop("Country", axis=1)
```

```python
features = dataset.iloc[:, 0:-1]
labels = dataset.iloc[:, -1:]
display(features)
```

```python
display(labels)
```

```python
features = pd.get_dummies(features)
display(features)
```

```python
features_train, features_test, labels_train, labels_test = sklearn.model_selection.train_test_split(features, labels,
    train_size = 0.33, random_state = 5)

labels_train = pd.DataFrame(labels_train)
labels_test = pd.DataFrame(labels_test)

display(features_train)
```

```python
numerical_features = features.select_dtypes(include=['float64', 'int64'])
numerical_columns = numerical_features.columns

ct = sklearn.compose.ColumnTransformer([("abc", sklearn.preprocessing.StandardScaler(), numerical_columns)],
                                       remainder = "passthrough")

features_test_scaled = pd.DataFrame(ct.fit_transform(features_test))
features_train_scaled = pd.DataFrame(ct.fit_transform(features_train))

display(features_test_scaled)
display(features_train_scaled)
```

```python
def build_model():
    m = tf.keras.models.Sequential([
        tf.keras.Input(shape=(features.shape[1],)),
        layers.Dense(30, "relu"), # Squishifies output according to relu
        layers.Dense(1) # No squishification
    ])

    opt = tf.keras.optimizers.legacy.Adam(learning_rate=0.01)
    m.compile(optimizer=opt, loss="mse", metrics=["mae"])

    return m

def train_model(m:tf.keras.models.Sequential):
    m.fit(features_train_scaled, labels_train, epochs=40, batch_size=1, verbose=1)

model = build_model()
print(model.summary())
```

```python
train_model(model)
res_mse, res_mae = model.evaluate(features_test_scaled, labels_test, verbose=0)
print("res_mse: " + str(res_mse)+ "\nres_mae: " + str(res_mae))
```
